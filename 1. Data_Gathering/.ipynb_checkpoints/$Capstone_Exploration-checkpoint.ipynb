#### Brick Economy - 01-12-2021

```python
brick_economy = pd.read_csv('data_gathering/brick_econ_overview01112021.csv')
brick_economy
```

```python
#Check for nulls
brick_economy.isna().mean()
```

```python
# Drop the last two columns with a large portion of null values
brick_economy = brick_economy.drop(['Available at', 'Not yet'], axis =1)

# Rename columns
brick_economy = brick_economy.rename(columns = 
        {'Retail': 'rrp', 'Value': 'value', 'Growth': 'growth', 'Annual growth':'annual_growth'})
```

```python
# Extract $ value from rrp and value columns
brick_economy['rrp'] = brick_economy['rrp'].str.extract('(\d+)')

brick_economy['value'] = brick_economy['value'].str.extract('(\d+)')
```

```python
# Extract % value from growth and annual_growth columns
brick_economy['growth'] = brick_economy['growth'].str.extract('(\d+)')

brick_economy['annual_growth'] = brick_economy['annual_growth'].str.extract('(\d+)')
```

```python
brick_economy
```

```python
brick_economy.to_csv('brick_economy01122021.csv', index = False)
```

#### Bricklink

```python
bricklink1 = pd.read_csv('Archive/bricklink_price.csv', index_col = 0, header =0, names = ['set_num', 'value_new', 'value_used'])
bricklink2 = pd.read_csv('Archive/bricklink_price2.csv', index_col = 0, header =0, names = ['set_num', 'value_new', 'value_used'])
```

```python
bricklink_price_01_11_2021 = pd.concat([bricklink1,bricklink2], axis =0, ignore_index = True).\
drop_duplicates(ignore_index = True)
```

```python
bricklink_price_01_11_2021['value_new'] = bricklink_price_01_11_2021['value_new'].str.split('$').str[1]

bricklink_price_01_11_2021['value_used'] = bricklink_price_01_11_2021['value_used'].str.split('$').str[1]
```

```python
bricklink_price_01_11_2021.to_csv('bricklink_price_01_11_2021.csv', index = False)
```
